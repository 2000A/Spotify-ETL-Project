```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
clientId = os.getenv('CLIENT_ID')
secretId = os.getenv('SECRET_ID')
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=clientId, client_secret=secretId)
```

```python
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/6SAUxRqgv0EdwHHnjBRmoE"
```

```python
playlist_URI = playlist_link.split("/")[-1].split('?')[0]
```

```python
playlist_URI
```

```python
sp.playlist(playlist_URI)
```

```python
sp.playlist_tracks(playlist_URI)
```

```python
data = sp.playlist_tracks(playlist_URI)
```

```python
album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    #album_band_name = row['track']['album']['artists'][0]['name']
    album_element = {'album_id':album_id, 'name':album_name, 'release_date':album_release_date,
                     'total_tracks':album_total_tracks,'url':album_url}
    album_list.append(album_element)


    #print(album_element)
    #print(album_band_name)
```

```python
album_list
```

```python
data['items'][23]['track']['album']['artists']
```

```python
artist_list = []
for row in data['items']:
    for key, value in row.items():
        if key == "track":
            for artist in value['album']['artists']:
                artist_dict = {'artist_id':artist['id'], 'artist_name':artist['name'],
                               'external_url':artist['href']}
                artist_list.append(artist_dict)
```

```python
artist_list
```

```python
album_df = pd.DataFrame.from_dict(album_list) #from_dict() pandas method creates a DataFrame from a python dictionary
```

```python
album_df
```

```python
album_df.info()  #At the time of transformation we will make release_date in proper datetime format
```

```python
album_df = album_df.drop_duplicates(subset=['album_id'])
```

```python
artist_df = pd.DataFrame.from_dict(artist_list)
```
